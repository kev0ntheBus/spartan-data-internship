```python
# Variables: used to store data
name = "Kevin"
age = 31
skills = ["python", "sql", "boxing"]
is_ready = True
```

```python
# Data Types: the function type() we use here to show that all data have types that categorize them
print(type(name), type(age), type(skills), type(is_ready))
```

```python
# Lists, which are another data type, as well as their their methods(functions built-in specific to a said data type). Also using Python functions on them as well
numbers = [10, 20, 30, 40, 50]

numbers.append(60)
numbers.remove(20)
numbers.sort()

print(min(numbers))
print(max(numbers))
print(sum(numbers))
```

```python
# Dctionaries, a powerful data structure roughly equivalent to Hashmaps, minus some smaller differences. Use key, value pairs and work insertion order
# Also touch on the idea of loops using for loops 
employee = {
    "name": "Kevin",
    "role": "Data Intern",
    "salary": 0,
    "skills": ["Python", "SQL", "Boxing"]
}

print(employee.keys())
print(employee.values())

employee["languages"] = ["English", "Spanish", "Tagalog", "Indonesian"]

for skill in employee["skills"]:
    print("Skills: ", skill)

for language in employee["languages"]:
    print("Languages: ", language)
```

```python
# Conditionals, which work when a condition is true, touched on by if else statements 
if age > 25:
    print("You are an adult intern.")
else:
    print("You are a junior intern.")
```

```python
# Functions: reusable blocks of code that are also already found in all programming langugaes with many built-in functions or methods(Class/Object-specific functions)
def mean(numbers):
    return sum(numbers) / len(numbers)

print(mean([10, 20, 30]))
```

```python
# File I/O or Input/Output: How one reads from and writes to files
with open("./test_data.txt", "w") as f:
    f.write("This is a test file created by Kevin the Data Intern")

with open("./test_data.txt", "r") as f:
    print(f.read())
```
